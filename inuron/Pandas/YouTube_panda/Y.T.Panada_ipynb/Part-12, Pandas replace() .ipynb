```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\ADMIN\\Desktop\\Y.T.practice\\Fortune_10 _09.csv')
df
```

```python
df.replace(to_replace = 'Health', value = 'Finance')   # 'Health' IS REPLACE IN 'Finance'
```

```python
df.replace('Health', 'Finance')   # 'Health' IS REPLACE IN 'Finance'  ( both same...without parameter)
```

```python
# change 'int' value

df.replace(10 , 15)   # 10 change 15(id chnge)
```

```python
df.replace([1,2,3,4,5,6,7,8], 0 )   # more then one value change..
```

```python
df.replace([1,2,3,4,5,6,7,8], [11,12,13,14,15,16,17,18] )   # more then one value change..
```

```python
df.replace({'Industry':'Health'},'none')   # ..columns value change..
```

```python
# ..columns value change..( more then one columns..che with one value)


df.replace({'Industry':'Health', 'Profit': 0 },'none')  
```

```python
# regex ...  'string' value replace with 'int' values...

df
```

```python
df.replace('[A-Za-z]', 0)    # no change..
```

```python
df.replace('[A-Za-z]', 0, regex = True)    # 'string' value replace with 'int' values...
```

```python
# spacific columns use ( only 'industry' columns )

df.replace({'Industry':'[A-Za-z]'}, 0, regex = True)
```

```python
# method..

df
```

```python
df.replace('Health', method = 'ffill')    # hear 'Health' values ... previous rows value fill
```

```python
df.replace('Health', method = 'bfill')   # hear 'Health' values ... backword rows value fill
```

```python
df.replace( 0, method = 'bfill', limit = 1 )       # hear '0' values ... backword rows value fill ( only one)
```

```python
df.replace( 0, method = 'bfill', limit = 2 )       # hear '0' values ... backword rows value fill ( only two)
```

```python
# modify table

df
```

```python
df.replace( 0, 100, inplace = True)    # '0' inplace 100
df
```
