```python
df1 = pd.read_csv('C:\\Users\\ADMIN\\Desktop\\Y.T.practice\\Fortune_5(1)_1.csv')  # all missing rows value shows 'NaN'
df1
```

```python
df1.dropna( how = 'all')           # all 'nan' rows drop..
```

```python
# Thresh
```

```python
df1
```

```python
# (thresh = 1) mins only one 'not null' or 'without null' values rows shows, and more then one null rows drope..(4. rows drope..)

df1.dropna(thresh = 1) 
```

```python
df1.dropna(thresh = 2) # every coloums has more then two not null values , that ways not change hear
```

```python
df1.dropna(thresh = 4) # id.4 ..has 3 not null values , that ways last rows(id.4 ) drope..
```

```python
# Subset..
```

```python
df1
```

```python
df1.dropna(subset = ['ID'])  # in 4 has nan values that ways drope.
```

```python
df1.dropna(subset = ['Name',])  # in 3 and 4 has nan values that ways drope.
```

```python
df1.dropna(subset = ['Profit'])  # in 4 has nan values that ways last rows drope.
```

```python
df1
```

```python
df1.dropna(subset = ['id','Name',])   
```

```python
# inplace.. 
```

```python
df1
```

```python
df1.dropna(inplace = True)   # all 'nan vlaues' rows drope..(new data frame not creat, but same data frame update..)
df1
```
