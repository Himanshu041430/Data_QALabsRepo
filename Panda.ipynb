### 1. Display all columns that contain any null values:

```python
# Get columns with any NaN values
null_columns = df2.columns[df2.isnull().any()]

# Show the columns with NaN values
print(null_columns)
```

### 2. Display all rows that contain any null values: 

```python
# Get rows with any NaN values
null_rows = df2[df2.isnull().any(axis=1)]

# Show rows with NaN values
print(null_rows)
```

### 3. Display the count of null values in each column:

```python
# Count of NaN values in each column
null_counts = df2.isnull().sum()

# Show the counts of NaN values per column
print(null_counts)
```

### 4. Display specific columns with null values:

```python
# Get a subset of columns with NaN values
df_null_columns = df2[null_columns] == True

# Show DataFrame with columns that have NaN values
print(df_null_columns)
```
